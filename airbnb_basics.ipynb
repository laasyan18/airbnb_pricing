```python
import pandas as pd
import numpy as np
df = pd.read_csv("Airbnb_Open_Data.csv")
print(df.head())
print(df.tail())
```

```python
df['price'] = df['price'].str.replace(r'[\$,]', '', regex=True).str.strip()

df['price'] = pd.to_numeric(df['price'], errors='coerce')

df = df.dropna(subset=['price'])

df['is_expensive'] = np.where(df['price'] > 300, True, False)

print(df[['price', 'is_expensive']].head())
print("Total rows after cleaning:", len(df))
```

```python
print(df.isnull().sum())
df['reviews per month'] = df['reviews per month'].fillna(0)
```

```python
print(df['minimum nights'].describe())
# You can remove listings with unrealistic minimum nights:
print(df[df['minimum nights'] < 30])
```

```python
print("Mean price:", df['price'].mean())
print("Median price:", df['price'].median())
print("Price standard deviation:", df['price'].std())
```

```python
print(df.groupby('neighbourhood group')['price'].mean())
```
